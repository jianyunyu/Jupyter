## Introduction

This is a short tutorial for implementation Discontinuous Galerkin (DG) methods for the advection equation,
\begin{align}
    u_t + u_x & = 0,\quad x \in [0,1] \\
    u(x,0) & = u_0(x) 
\end{align}
with periodic boundary conditions.
***

## The Step by Step Guide

### The first step
First of all, we need make a partition of the domain $[0,1]$,
$$0 = x_{\frac{1}{2}} < x_{\frac{3}{2}} < \cdots < x_{N+\frac{1}{2}},$$
where 
- the element/cell $I_j = [x_{j-\frac{1}{2}}, x_{j+\frac{1}{2}}]$
- the middle point of the $j$th element, $x_j = \frac{1}{2}(x_{j-\frac{1}{2}} + x_{j+\frac{1}{2}})$
- the size of the $j$th element, $\Delta x_j = x_{j+\frac{1}{2}} - x_{j-\frac{1}{2}}$

To store the mesh, one way is to store the middle point $x_j$ and the size $\Delta x_j$

```python
import numpy as np
import numpy.polynomial.legendre as npleg
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import clear_output, display
```

```python
def generate_mesh(left, right, N):
    # uniform mesh
    x = np.zeros(N+1)
    h = (right - left)/N
    for j in range(1,N+1):
        x[j] = (j - 0.5)*h
    
    dx = np.zeros(N+1) + h
    return  x, dx
x, dx = generate_mesh(0, 1, 10)
```

### The DG (finite element) space
Now, we need to define the finite element space 
$$ V_h^k = \left\{v: v|_{I_j} \in \mathbb{P}^k(I_j),\, 1 \leq j \leq N\right\},$$
where $\mathbb{P}^k(I_j)$ denotes the set of polynomials of degree up to $k$ defined on the element $I_j$.  

For $v \in V_h^k$, 
- piecewise polynomials 
- **discontinous** across the element interface (bi-values)

To implement the finite element space $V_h^k$, we just need to choose a basis of $\mathbb{P}^k$, for example,
- monomial basis, $1, x, x^2, \ldots, x^k$,
- Legendre polynomial on $[-1,1]$,
$$ P_n(x) = \frac{1}{2^n n!}\frac{d^n}{dx^n}\left((x^2-1)^n\right), \quad x \in [-1,1].$$
    - orthogonality
        $$\int_{-1}^1 P_\ell(x)P_m(x) = \frac{2}{2\ell+1}\delta_{\ell,m}$$
    - recursive relation
        $$ (n+1)P_{n+1}(x) = (2n+1)xP_n(x) - n P_{n-1}(x)$$
    - end points
        $$ P_\ell(\pm1) = (\pm)^\ell $$
    - first several polynomials:
        $$P_0(x) = 1,\,\, P_1(x) = x,\,\, P_2(x) = \frac{1}{2}(3x^2-1), P_3(x)= \frac{1}{2}(5x^2-3),\,\,\ldots$$

In this tutorial, we are going to use the Legendre polynomial to construct the basis of the DG space, for element $I_j$, 
$$ \varphi_j^\ell(x) = P_\ell\left(\frac{2(x - x_j)}{\Delta x_j}\right),\quad x \in I_j$$

```python
def LegendreBasis(x, n):
    if (n == 0):
        return 1.
    elif (n == 1):
        return x
    elif (n == 2):
        return 1.5*x*x - 0.5
    elif (n == 3):
        return 0.5*x*(5*x*x - 3)
    else:
        raise Exception("not yet")
```

### The DG formulation

Now, we are ready to implement the DG formulation.  That is find the numerical solution $u_h \in V_h^k$, such that, on element $I_j$, 
$$ \int_{I_j}(u_h)_t v_h \,dx - \int_{I_j}u_h(v_h)_x\,dx + \hat{u}_h(x_{j+\frac{1}{2}})v_h(x_{j+\frac{1}{2}}^{-}) - \hat{u}_h(x_{j-\frac{1}{2}})v_h(x_{j-\frac{1}{2}}^{+}) = 0,\quad\, \text{for}\,\forall v_h \in V_h^k,$$
where the numerical flux $\hat{u}_h(x_{j+\frac{1}{2}})$ is chosen as the upwind flux $u_h(x_{j+\frac{1}{2}}^{-})$. 

There are three terms need to be taken care: 
1. $\int_{I_j}(u_h)_t v_h \,dx$
2. $\int_{I_j}u_h(v_h)_x\,dx$
3. $\hat{u}_h(x_{j+\frac{1}{2}})v_h(x_{j+\frac{1}{2}}^{-}) - \hat{u}_h(x_{j-\frac{1}{2}})v_h(x_{j-\frac{1}{2}}^{+})$ 

To be continue, let us write the numerical solution $u_h$ with its modal form
$$ u_h(x, t) = \sum\limits_{\ell = 0}^k u_j^\ell(t)\varphi_j^\ell(x),\quad x \in I_j,$$
and the test function 
$$ v_h = \varphi_j^0, \varphi_j^1, \ldots, \varphi_j^k.$$

***
**Term 1**

\begin{align}
    \int_{I_j} (u_h)_t v_h\,dx &= \int_{I_j} \sum\limits_{\ell=0}^k\dot{u}_j^\ell(t)\varphi_j^\ell(x)\varphi_j^m(x)\, dx\qquad m = 0,\ldots,k \\
                               &= \sum\limits_{\ell=0}^k\dot{u}_j^\ell(t)\int_{I_j}\varphi_j^\ell(x)\varphi_j^m(x)\, dx \\
                               &= \sum\limits_{\ell=0}^k\dot{u}_j^\ell(t)\left(\varphi_j^\ell(x)\varphi_j^m(x)\right)_j
\end{align}

Denote the unknowns 
$$U_j(t) = \begin{bmatrix}u_j^0(t) \\ u_j^1(t) \\ \vdots \\ u_j^k(t)\end{bmatrix},$$
and the mass matrix
$$
    M_j = \begin{bmatrix}\left(\varphi_j^0, \varphi_j^0\right)_j & \cdots & \left(\varphi_j^k, \varphi_j^0\right) \\ 
    \vdots &\ddots & \vdots \\
    \left(\varphi_j^0, \varphi_j^k\right)_j &\cdots & \left(\varphi_j^k, \varphi_j^k\right)\end{bmatrix}.
$$

In fact, since we define the basis $\varphi_j^\ell$ based on the Legendre polynomial $P_\ell$ on the reference element $[-1,1]$, the entry of the mass matrix 
\begin{align}
    \left(\varphi_j^\ell,\varphi_j^m\right) & = \int_{I_j}\varphi_j^\ell(x)\varphi_j^m(x)\, dx = \int_{I_j}P_\ell\left(\frac{2(x - x_j)}{\Delta x_j}\right)P_m\left(\frac{2(x - x_j)}{\Delta x_j}\right)\, dx \\
    & = \int_{-1}^{1}P_\ell(\xi)P_m(\xi)\frac{\Delta x_j}{2}\,d\xi \\
    & = \frac{\Delta x_j}{2}\frac{2}{2\ell+1}\delta_{\ell,m}. 
\end{align}
This leads the mass matrix
$$
    M_j = \frac{\Delta x_j}{2}M = \frac{\Delta x_j}{2}\begin{bmatrix}2 & & &  \\
     & \frac{2}{3} &  & \\
    & &\ddots &   \\
    &  &  & \frac{2}{2k+1}\end{bmatrix}.
$$
That is 
$$
\int_{I_j} (u_h)_t v_h\,dx = \frac{\Delta x_j}{2}M \dot{U}_j(t)
$$
*Note:* in fact, we only need the inverse of the mass matrix,
$$
M^{-1} = \begin{bmatrix}\frac{1}{2} & & &  \\
     & \frac{3}{2} &  & \\
    & &\ddots &   \\
    &  &  & \frac{2k+1}{2}\end{bmatrix}.
$$

```python
def generate_mass_matrix(k):
    M = np.diag([2/(2*i+1) for i in range(k+1)])
    M_inv = np.diag([(2*i+1)/2. for i in range(k+1)])
    return M_inv,  M
```

***
**Term 2**

\begin{align}
    \int_{I_j}u_h(v_h)_x\,dx &= \int_{I_j} \sum\limits_{\ell=0}^k u_j^\ell(t)\varphi_j^\ell(x)\left(\varphi_j^m(x)\right)_x\, dx\qquad m = 0,\ldots,k \\
                               &= \sum\limits_{\ell=0}^k u_j^\ell(t)\int_{I_j}\varphi_j^\ell(x)\dot{\varphi}_j^m(x)\, dx \\
                               &= \sum\limits_{\ell=0}^k u_j^\ell(t)\left(\varphi_j^\ell(x)\dot{\varphi}_j^m(x)\right)_j
\end{align}

Here, we denote the matrix
$$
    S_j = \begin{bmatrix}\left(\varphi_j^0, \dot{\varphi}_j^0\right)_j & \cdots & \left(\varphi_j^k, \dot{\varphi}_j^0\right) \\ 
    \vdots &\ddots & \vdots \\
    \left(\varphi_j^0, \dot{\varphi}_j^k\right)_j &\cdots & \left(\varphi_j^k, \dot{\varphi}_j^k\right)\end{bmatrix}.
$$
as the stiff matrix.

Similar as the mass matrix, we have 
\begin{align}
    \left(\varphi_j^\ell,\dot{\varphi}_j^m\right) & = \int_{I_j}\varphi_j^\ell(x)\dot{\varphi}_j^m(x)\, dx = \int_{I_j}P_\ell\left(\frac{2(x - x_j)}{\Delta x_j}\right)\dot{P}_m\left(\frac{2(x - x_j)}{\Delta x_j}\right)\frac{2}{\Delta x_j}\, dx \\
    & = \int_{-1}^{1}P_\ell(\xi)\dot{P}_m(\xi)\,d\xi. \\ 
\end{align}
We note that the stiff matrix is not dependent on $j$, namely $S_j = S$. Usually, one can calculate it by using Gaussian quadrature.  However, we are using the Legendre polynomials, for $m > \ell$
$$
\int_{-1}^{1}P_\ell(x)\dot{P}_m(x)\,dx = P_\ell(x)P_m(x)|_{-1}^1 - \int_{-1}^{1}\dot{P}_\ell(x) P_m(x)\,dx = 1 - (-1)^{\ell+m}.
$$
This leads 
$$
\int_{-1}^{1}P_\ell(\xi)\dot{P}_m(\xi)\,d\xi = \left\{\begin{array}{cc}0 & m\leq \ell \\ 1-(-1)^{\ell+m} & m > \ell \end{array}\right.
$$
That is 
$$
\int_{I_j} u_h \dot{v}_h\,dx = S U_j(t)
$$

```python
def generate_stiff_matrix(k):
    S = np.zeros([k+1,k+1])
    for i in range(k+1):
        for j in range(i+1, k+1, 2):
            S[j,i] = 2
    return S
```

***
**Term 3**

Now, let us have a look at the final term, the flux term, which is always the most confussion part of the DG methods.
\begin{align}
 & \,\hat{u}_h(x_{j+\frac{1}{2}})v_h(x_{j+\frac{1}{2}}^{-}) - \hat{u}_h(x_{j-\frac{1}{2}})v_h(x_{j-\frac{1}{2}}^{+}) \qquad m = 0,\ldots, k \\ 
 = &\, \sum\limits_{\ell=0}^k u_j^\ell(t)\varphi_j^\ell(x_{j+1}^{-})\varphi_j^m(x_{j+1}^{-}) - \sum\limits_{\ell=0}^k u_{j-1}^\ell(t)\varphi_{j-1}^\ell(x_{j-1}^{-})\varphi_j^m(x_{j-1}^{+}) \\
 = &\, \sum\limits_{\ell=0}^k u_j^\ell(t) - \sum\limits_{\ell=0}^k u_{j-1}^\ell(t)(-1)^m 
\end{align}

For convenience, let us define the flux matrice,
$$
F_r = \begin{bmatrix}1 & 1 & \cdots & 1 \\ 1 & 1 & \cdots & 1 \\ \vdots & \vdots & \ddots & \vdots \\ 1 & 1 & \cdots & 1\end{bmatrix},\qquad
F_l = \begin{bmatrix}1 & 1 & \cdots & 1 \\ -1 & -1 & \cdots & -1 \\ \vdots & \vdots & \ddots & \vdots \\ (-1)^k & (-1)^k & \cdots & (-1)^k\end{bmatrix}.
$$
That is 
$$
\hat{u}_h(x_{j+\frac{1}{2}})v_h(x_{j+\frac{1}{2}}^{-}) - \hat{u}_h(x_{j-\frac{1}{2}})v_h(x_{j-\frac{1}{2}}^{+}) = F_r U_j(t) - F_l U_{j-1}(t).
$$

```python
def generate_flux_matrix(k):
    Fr = np.ones([k+1,k+1])
    Fl = np.ones([k+1,k+1])
    for i in range (1, k+1, 2):
        Fl[i,:] = -Fl[i,:]
    return Fr, Fl
```

***
**Semi-Discrete DG formulation**

Finally, by putting all three terms together, we have the semi-discrete DG formulation,
$$
\frac{\Delta x_j}{2}M \dot{U}_j(t) = S U_j(t) - Fr U_j(t) + F_l U_{j-1}(t), \qquad j = 1,\ldots, N.
$$
For the boundary flux $U_{0}$, it is decided by the boundary condition:
- for periodic boundary condition, $U_0(t) = U_N(t)$,
- for Dirichlet boundary condition, $U_0(t) = u_l(t)$.

### Time Discretization

**The Third Order TVD Runge-Kutta Method**

The most popular scheme used for semi-discrete DG scheme is the following third order Runge-Kutta method for solving 
$$ u_t = \text{Res}(u, t),$$
where $\text{Res}(u,t)$ is the spatial discretization operator
\begin{align}
    u^{(1)} & = u^n + \Delta t \text{Res}(u^n, t^n) \\
    u^{(2)} & = \frac{3}{4}u^n + \frac{1}{4}u^{(1)} + \frac{1}{4}\Delta t \text{Res}(u^{(1)}, t^n + \Delta t) \\
    u^{n+1} & = \frac{1}{3}u^n +\frac{2}{3}u^{(2)} + \frac{2}{3}\Delta t \text{Res}(u^{(2)}, t^n+\frac{1}{2}\Delta t).
\end{align}

***
Now, we couple the DG spatial discretization with the third ordre Runge-Kutta method (or other explicit time discretizations).  When coupled with explicit time discretization, we can solve the entire system element by element (or say block by block).  

First of all, let us introduce the residue notation to simplify the writing
$$
    \dot{U}_j(t) = \text{Res}(U_j(t), U_{j-1}(t)), \qquad j = 1,\ldots, N.
$$
where $\text{Res}(U_j(t), U_{j-1}(t)) = \frac{2}{\Delta x_j}M^{-1}\left(S U_j(t) - Fr U_j(t) + F_l U_{j-1}(t)\right)$.

```python
def Residue(u, dx, N, M_inv, S, Fr, Fl):
    Res = np.zeros(np.shape(u))
    k = np.size(u, 1)-1
    
    # periodic boundary 
    u[0,:] = u[N,:]
    
    for j in range(1, N+1):
        Res[j,:] = 2/dx[j]*np.dot(M_inv, np.dot(S,u[j,:]) - np.dot(Fr, u[j,:]) + np.dot(Fl, u[j-1,:]))
  
    return Res
```

***
For the first stage in the above Runge-Kutta method (or say the forward Euler method),
- for $j = 1$ to $N$:
$$
    U_j^{(1)} = U_j^n + \Delta t \text{Res}(U_j^n, U_{j-1}^{n})
$$
    - where $U_j^n = U_j(t^n)$  
    - note for periodic boundary conditions, $U_0^{n} = U_N^n$.
   
For the second stage,
- for $j = 1$ to $N$:
$$
    U_j^{(2)} = \frac{3}{4}U_j^n + \frac{1}{4}U_j^{(1)} + \frac{1}{4}\Delta t \text{Res}(U_j^{(1)}, U_{j-1}^{(1)}).
$$

For the third stage,
- for $j = 1$ to $N$:
$$
    U_j^{n+1} = \frac{1}{3}U_j^n + \frac{2}{3}U_j^{(2)} + \frac{2}{3}\Delta t \text{Res}(U_j^{(2)}, U_{j-1}^{(2)}).
$$

*Note:*
- for some equation, $\text{Res}$ may also depend on time $t$. 

```python
def Runge_Kutta(t, dt, u, dx, N, M_inv, S, Fr, Fl):
    # The thrid order Runge-Kutta method
    
    # save the u^n data
    u_n = np.copy(u)
    
    # first stage
    Res = Residue(u, dx, N, M_inv, S, Fr, Fl)
    u = u_n + dt*Res
    
    # second stage
    Res = Residue(u, dx, N, M_inv, S, Fr, Fl)
    u = 0.75*u_n + 0.25*u + 0.25*dt*Res
    
    # third stage
    Res = Residue(u, dx, N, M_inv, S, Fr, Fl)
    u = (u_n + 2*u + 2*dt*Res)/3.
    
    return u
```

```python
def DG_time(u, x, dx, cfl, time, N, M_inv, S, Fr, Fl):
    dt = cfl*np.max(dx)
    t =  dt
    while (t < time):
        u = Runge_Kutta(t, dt, u, dx, N, M_inv, S, Fr, Fl)
        t = t + dt
    dt = time + dt - t
    u = Runge_Kutta(time, dt, u, dx, N, M_inv, S, Fr, Fl)
    return u
```

***
**Initial Time**

From the above section, we know that how to march in time through the Runge-Kutta method.  However, we still need to know the initial data $U_j^0 = U_j(0)$.  In this tutorial, we will do a simple $L^2$ projection.  For element $I_j$, we have
\begin{align}
\int_{I_j} u_h v_h \,dx & = \int_{I_j} u_0 v_h\, dx \\
\int_{I_j}\sum\limits_{\ell=0}^k u_j^\ell(0)\varphi_j^\ell(x) \varphi_j^m(x)\, dx &= \int_{I_j}u_0(x) \varphi_j^m(x)\, dx \\
\sum\limits_{\ell=0}^k u_j^\ell(0)\frac{\Delta x_j}{2}\int_{-1}^1 P_\ell(\xi) P_m(\xi)\,d\xi & = \frac{\Delta x_j}{2}\int_{-1}^1 u_0\left(x_j + \frac{\Delta x_j}{2}\xi\right)P_m(\xi)\,d\xi \\ 
\sum\limits_{\ell=0}^k u_j^\ell(0)\int_{-1}^1 P_\ell(\xi) P_m(\xi)\,d\xi & = \int_{-1}^1 u_0\left(x_j + \frac{\Delta x_j}{2}\xi\right)P_m(\xi)\,d\xi
\end{align}
That is,
$$
    U_j(0) = M^{-1}F_j,\qquad j = 1,\ldots,N
$$
where
$$
    F_j =  \begin{bmatrix}\int_{-1}^1 u_0\left(x_j + \frac{\Delta x_j}{2}\xi\right)P_0(\xi)\,d\xi \\ \int_{-1}^1 u_0\left(x_j + \frac{\Delta x_j}{2}\xi\right)P_1(\xi)\,d\xi \\ \vdots \\ \int_{-1}^1 u_0\left(x_j + \frac{\Delta x_j}{2}\xi\right)P_k(\xi)\,d\xi\end{bmatrix},\qquad j = 1,\ldots,N
$$
can be calculated by using numerical quadrature, for example, the Gaussian quadrature.

```python
def initial_projection(fun0, k, N, x, dx, M_inv):
    # Gaussian quadrature
    #Gpn = int(np.ceil(0.5*(num+order)))
    Gpn = 6
    xg, wg = npleg.leggauss(Gpn)
    u =  np.zeros([N+1,k+1])
    
    F = np.zeros(k+1)
    for j in range(1, N+1):
        for l in range(k+1):
            F[l] = 0.
            for m in range(Gpn):
                F[l] += wg[m]*fun0(x[j] + 0.5*dx[j]*xg[m])*LegendreBasis(xg[m], l)
        u[j,:] = np.dot(M_inv, F)
    return u
```

***
### The DG solver for Advection Equation.

Now, let us put every thing together, 
1. we set a function to genearte the mesh and all the matrice. 

```python
def DG_init(k, N, left, right):
    # generate the mesh
    x,  dx = generate_mesh(left, right, N)
    
    # generate all the matrice
    M_inv, _ = generate_mass_matrix(k)
    S = generate_stiff_matrix(k)
    Fr, Fl = generate_flux_matrix(k)
    
    return x, dx, M_inv, S, Fr, Fl
```

2. we define the initial function 
$$
    u_0(x) = \sin(2\pi x),
$$
and the exact solution is just
$$
    u(x,t) = u_0(x-t).
$$

```python
def fun0(x):
    return np.sin(2.*np.pi*x)
```

3. we set the DG solver for the advection equation $u_t + u_x = 0$.

```python
def DG_solver(k, N, left, right, fun0, time):
    # get the mesh, matrice
    x, dx, M_inv, S, Fr, Fl = DG_init(k, N, left, right)
    
    # set the cfl number
    cfl = 0.1**k
    
    # initial condition
    u = initial_projection(fun0, k, N, x, dx, M_inv)
    
    # time forward
    u = DG_time(u, x, dx, cfl, time, N, M_inv, S, Fr, Fl)
    
    return u, x, dx
```

```python
u, x, dx = DG_solver(1, 20, 0, 1, fun0, 1)
```

***
### Verification 

Now, we algready have a working DG code.  However, for mathematicians, we want to make sure the code does the expected job.

* **check error:** $\|u_h - u\|$

```python
def DG_error(u, x, dx, fun0, time):
    # set Gaussian quadrature to calculate the L2 norm 
    Gpn = 6
    xg, wg = npleg.leggauss(Gpn)
    err2 = 0.
    erri = 0.
    
    N = np.size(u, 0) - 1
    k = np.size(u, 1) - 1
    
    ## plotting
    plot_x = np.empty(N*Gpn)
    plot_y = np.empty(N*Gpn)
    
    for j in range(1,N+1):
        for m in range(Gpn):
            uh = 0.
            xj = x[j] + dx[j]*xg[m]/2
            for l in range(k+1):
                uh += u[j,l]*LegendreBasis(xg[m], l)
            err = np.abs(fun0(xj-time) - uh)
            err2 += 0.5*dx[j]*wg[m]*err**2
            erri = max(erri, err)
                
            plot_x[(j-1)*Gpn+m] = xj
            # plot_y[(j-1)*Gpn+m] = uh
            plot_y[(j-1)*Gpn+m] = err
    err2 = np.sqrt(err2/1.)
    plt.plot(plot_x, plot_y, linewidth=2.0)
    print ("N = ", N, "  k = ", k)
    print ("L2 error:   %7.2e" % err2)
    print ("Linf error: %7.2e" % erri)
    plt.yscale('log')
    plt.xlim(0, 1)
    #plt.ylim(1.e-6,1.e0)
    plt.xlabel('x',fontsize=18)
    plt.ylabel('|error|',fontsize=18)
    plt.show()
```

Now, let us have a look at the error of our numerical solution in the $L^2$ and $L^\infty$ norm, also we plot the point-wise error.

```python
DG_error(u, x, dx, fun0, 1)
```

From the above error and error, we can see that our numerical solution is a good approximation to the exact solution (well, in some sense).  However, it is not sufficient to show that the implementation is correct.  To further verify the code, we need to check the accuracy order.  That is, how the error changes along when the element number $N$ (the polynomial degree $k$).  In theory, the DG solution will have the optimal convergence rate of order $k+1$.  
* **check the accuracy order:** $\|u_h - u\| = \mathcal{O}(h^{k+1})$

First, in order to output a result in a good table form, we define a print table function to control the output format of of the results.
- one can also design a function to output results in a form that can be used directly in the Latex. 

```python
def print_OrderTable(number_coarse, erri, err2):
    print ("  N  Linf norm order  L2 norm  order")
    for i in range(np.size(erri)):
        N = number_coarse*2**i
        if (i == 0):
            print ("%3d  %7.2e   --   %7.2e   --" % \
                   (N, erri[i], err2[i]))
        else:
            print ("%3d  %7.2e  %4.2f  %7.2e  %4.2f" % \
                       (N, \
                        erri[i], np.log(erri[i-1]/erri[i])/np.log(2),\
                        err2[i], np.log(err2[i-1]/err2[i])/np.log(2)))
```

Now, let us check the accuracy order
- here, we fix the polynomial $k$, and change the number of element $N$.  We double $N$ in each level, then the error should be just $\frac{1}{2^{k+1}}$ (or taking $\log_2$ gives the $k+1$) of the previous level.  

```python
def DG_order(number_coarse, degree):
    Gpn = 6
    xg, wg = npleg.leggauss(Gpn)
    err2 = np.zeros(4)
    erri = np.zeros(4)
    style = ['b.','g-.','r--','c-']
    
    for i in range(4):
        N = number_coarse*2**(i)
        
        time = 1.
        left = 0.
        right = 1.
        
        u, x, dx = DG_solver(degree, N, left, right, fun0, time)
        
        ## plotting
        plot_x = np.empty(N*Gpn)
        plot_y = np.empty(N*Gpn)
        
        for j in range(1,N+1):
            for m in range(Gpn):
                uh = 0.
                xj = x[j] + dx[j]*xg[m]/2
                for l in range(degree+1):
                    uh += u[j,l]*LegendreBasis(xg[m], l)
                err = np.abs(fun0(xj-time) - uh)
                err2[i] += 0.5*dx[j]*wg[m]*err**2
                erri[i] = max(erri[i], err)
                
                plot_x[(j-1)*Gpn+m] = xj
                plot_y[(j-1)*Gpn+m] = uh
                plot_y[(j-1)*Gpn+m] = err
        err2[i] = np.sqrt(err2[i]/1.)
        plt.plot(plot_x, plot_y, style[i],linewidth=2.0, label="N = " + str(N))
    print_OrderTable(number_coarse, erri, err2)
    plt.yscale('log')
    ax = plt.legend(fontsize=14)
    ax.draw_frame(False)
    plt.xlim(0, 1)
    #plt.ylim(1.e-6,1.e0)
    plt.xlabel('x',fontsize=18)
    plt.ylabel('|error|',fontsize=18)
    plt.show()
DG_order(20, 2)
```

### Couple with Accuracy Enhance Filtering

From the above tutorial, we can see that the DG solution has accuracy order of $k+1$ in the $L^2$ norm, that is, 
$$
\| u - u_h\|_0 \leq C h^{k+1}.
$$
This is also the optimal convergence rate for using approximation polynomial of degree $k$ in the $L^2$ norm. 

In addition, one can use an accuracy enhace filtering (or, so-called Smoothness-Increasing Accuracy-Conserving (SIAC) filtering) technique.  By applying the filter technique, one can reconstruct the DG solution $u_h$ into a new filtered solution $u_h^*$ and improve the accuracy order from $k+1$ to $2k+1$,
$$    \|u - u_h^*\|_0 \leq C h^{2k+1}.  $$
For example, one can refer the following plot to see the improvment aftering applying the filtering technique.  Both the accuracy and the smoothness of the DG solution are greately improved.

### Implicit Integrator 

Of course, for the advection equation, one donot need (or say may not) to use the implicit time integrator.  The purpose of this section is to simply demonstrate how to implement the implicit integrator (or how to assemble the global matrix).  For this purpose, we simply consider the backward Euler method for the time discretiztion,  
$$
    u^{n+1}  = u^n + \Delta t \text{Res}(u^{n+1}, t^{n+1}).
$$


***
**Semi-Discrete DG formulation**

First, let us write down the semi-discrete DG formular,
$$
\frac{\Delta x_j}{2}M \dot{U}_j(t) = S U_j(t) - Fr U_j(t) + F_l U_{j-1}(t), \qquad j = 1,\ldots, N.
$$
For the boundary flux $U_{0}$, it is decided by the boundary condition:
- for periodic boundary condition, $U_0(t) = U_N(t)$,
- for Dirichlet boundary condition, $U_0(t) = u_l(t)$.


***
**Fully-Discrete formulation**

Now, we couple the DG spatial discretization with the backward Euler method
$$
    U^{n+1}_j(t) = U_j^n + 2\frac{\Delta t}{\Delta x_j}M^{-1}\left(S U^{n+1}_j(t) - Fr U^{n+1}_j(t) + F_l U^{n+1}_{j-1}(t)\right).
$$
Unlike coupling with explicit time discretization, the above system can not be solved element by element (or say block by block), one has to solve a global system.  Therefore, we need to rewrite the unknowns from 
$$U_j(t) = \begin{bmatrix}u_j^0(t) \\ u_j^1(t) \\ \vdots \\ u_j^k(t)\end{bmatrix},\qquad j = 1,\ldots, N,$$
as 
$$ U(t) = \begin{bmatrix}u_1^0(t) \\ u_1^1(t) \\ \vdots \\ u_1^k(t) \\ \vdots \\ u_N^0(t) \\ u_N^1(t) \\ \vdots \\ u_N^k(t)\end{bmatrix}. $$

***
**Global Matrix**
Then, we need to write the local formular 
$$
    2\frac{\Delta t}{\Delta x_j}M^{-1}\left(S U_j(t) - Fr U_j(t) + F_l U_{j-1}(t)\right)
$$
as global matrix form $\text{Res}(U(t)) = AU(t)$, where the matrix 
$$
    A = 2\Delta t\begin{bmatrix} \Delta x_1(S - F_r) &  &&& \Delta x_1 F_l\\
                                 \Delta x_2 F_l & \Delta x_2(S - F_r) & \\
                                                & \Delta x_3 F_l & \Delta x_3(S - F_r) & \\
                                                &                & \ddots & \ddots &  \\
                                                &                &        & \Delta x_N F_l & \Delta x_N(S - F_r)\end{bmatrix},
$$
or assume the uniform mesh $\Delta x = \Delta x_j$, 
$$
    A = 2\frac{\Delta t}{\Delta x}\begin{bmatrix} S - F_r &  &&& F_l\\
                                 F_l & S - F_r & \\
                                                & F_l & S - F_r & \\
                                                &                & \ddots & \ddots &  \\
                                                &                &        & F_l & S - F_r\end{bmatrix}.
$$

**Note:**
- In this example, oe can see that the global matrix $A$ is independent with the unknown $U$.  Therefore, in practice, there are many methods to efficiently solve the linear system.

```python
def Residue_global(dx, N, M_inv, S, Fr, Fl):
    k = np.size(S, 1)-1
    A = np.zeros([(N+1)*(k+1), (N+1)*(k+1)])
    
    for j in range(1, N+1):
        A[j*(k+1):(j+1)*(k+1), j*(k+1):(j+1)*(k+1)] = 1./dx[j]*np.dot(M_inv, S - Fr)
        if j > 1: 
            A[j*(k+1):(j+1)*(k+1), (j-1)*(k+1):j*(k+1)] = 1./dx[j]*np.dot(M_inv, Fl)
        else:
            A[j*(k+1):(j+1)*(k+1), N*(k+1):(N+1)*(k+1)] = 1./dx[j]*np.dot(M_inv, Fl)
    return A
```

***
Now, we solve the ODEs $\dot{U}(t) = A U(t)$ by the backward Euler method.

```python
def Implicit_Euler(t, dt, u, dx, N, A):
    # The thrid order Runge-Kutta method
    
    # save the u^n data
    #u_n = np.copy(u)
    
    u = np.linalg.solve(np.eye(np.size(u))-2*dt*A, u)
    
    return u
```

```python
def DG_time_imp(u, x, dx, cfl, time, N, A):
    dt = cfl*np.max(dx)
    t =  dt
    while (t < time):
        u = Implicit_Euler(t, dt, u, dx, N, A)
        t = t + dt
    dt = time + dt - t
    u = Implicit_Euler(time, dt, u, dx, N, A)
    return u
```

```python
def DG_solver(k, N, left, right, fun0, time):
    # get the mesh, matrice
    x, dx, M_inv, S, Fr, Fl = DG_init(k, N, left, right)
    
    # set the cfl number
    cfl = 0.001
    
    # initial condition
    u = initial_projection(fun0, k, N, x, dx, M_inv)
    # flatten the 2d arrary to 1d
    u = u.flatten()
    
    # global matrix
    A = Residue_global(dx, N, M_inv, S, Fr, Fl)
    

    # time forward
    u = DG_time_imp(u, x, dx, cfl, time, N, A)
    
    # let reshape the unknown back to 2d array, so we can use the verification functions
    u = np.reshape(u, [N+1,k+1])
    return u, x, dx
```
